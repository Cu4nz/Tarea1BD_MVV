## Crear tablas

Aquí inserta el código SQL que creara las tablas de la base de datos:

### Tabla "Director"

```python
%%sql
CREATE TABLE director(

    id_director INT PRIMARY KEY GENERATED ALWAYS AS IDENTITY,

    nombre VARCHAR(45) NOT NULL,

    edad INTEGER NOT NULL

);
```

### Tabla "Actor"

```python
%%sql

CREATE TABLE actor (

id_actor INT PRIMARY KEY GENERATED ALWAYS AS IDENTITY,

nombre VARCHAR(45) NOT NULL,

edad INTEGER NOT NULL

)

```

### Tabla "Pelicula"

```python
%%sql
CREATE TABLE pelicula (

    id_pelicula INT PRIMARY KEY GENERATED ALWAYS AS IDENTITY,

    id_director INT ,

    nombre VARCHAR(45) NOT NULL,

    genero VARCHAR(45) NOT NULL,

    duracion INTEGER NOT NULL,

    fecha_publicacion DATE NOT NULL,

    FOREIGN KEY (id_director) REFERENCES director(id_director)

);
```

### Tabla "actor_participa_en_pelicula"

```python
%%sql
CREATE TABLE actor_participa_en_pelicula (

    id_pelicula INT,

    id_actor INT,

    FOREIGN KEY (id_pelicula) REFERENCES pelicula(id_pelicula),

    FOREIGN KEY (id_actor) REFERENCES actor(id_actor)

);
```

### Tabla  "Premio_mejor_director"

```python
%%sql
CREATE TABLE premio_mejor_director (

    id_premio INT PRIMARY KEY GENERATED ALWAYS AS IDENTITY,

    id_director INT NOT NULL,

    anio INTEGER NOT NULL,

    FOREIGN KEY (id_director) REFERENCES director(id_director)

);
```

### Tabla  "Premio_mejor_actor"

```python
%%sql
CREATE TABLE premio_mejor_actor (

    id_premio INT PRIMARY KEY GENERATED ALWAYS AS IDENTITY,

    id_actor INT,

    anio INTEGER NOT NULL,

    FOREIGN KEY (id_actor) REFERENCES actor(id_actor)

);
```

### Tabla  "Premio_mejor_pelicula"

```python
%%sql
CREATE TABLE premio_mejor_pelicula (

    id_premio INT PRIMARY KEY GENERATED ALWAYS AS IDENTITY,

    id_pelicula INT,

    anio INTEGER NOT NULL,

    FOREIGN KEY (id_pelicula) REFERENCES pelicula(id_pelicula)

);
```

### Consulta 6
**Nombre de la película con la mayor brecha de edad entre el actor más joven y el actor más viejo que participen en ella, y las edades de dichos actores.** (Si hay empate, mostrar solamente uno).


**Ejemplo:**
|Nombre|Edad del actor más joven|Edad del actor más viejo|
|------|------------------------|------------------------|
|Barbie|16                      |92                      |


**Solución:** 
- **Cálculo de edades:** En la parte del código MIN(a.edad) se calcula la edad mínima de los actores y en MAX(a.edad) la máxima.
- **Relación de tablas:** La tabla pelicula se relaciona con actor_participa_en_pelicula mediante JOIN. Luego, actor_participa_en_pelicula se relaciona con actor también mediante JOIN.
- **Agrupación:** Se agrupa el resultado por el nombre de la película (GROUP BY p.nombre) para obtener las edades mínima y máxima por película.
- **Ordenamiento y limitación:** Se ordena el resultado por la diferencia entre la edad máxima y mínima (ORDER BY (MAX(a.edad) - MIN(a.edad)) DESC) y se limita a mostrar solo una fila (LIMIT 1), lo que devuelve la película con la mayor brecha de edad entre sus actores.

```python
%%sql
SELECT p.nombre AS nombre_pelicula, 
       brechas_edad_por_pelicula.edad_min AS edad_actor_mas_joven,
       brechas_edad_por_pelicula.edad_max AS edad_actor_mas_viejo
FROM pelicula p
JOIN (
    SELECT id_pelicula,
           MAX(a.edad) AS edad_max,
           MIN(a.edad) AS edad_min
    FROM actor_participa_en_pelicula ap
    JOIN actor a ON ap.id_actor = a.id_actor
    GROUP BY id_pelicula
) AS brechas_edad_por_pelicula ON p.id_pelicula = brechas_edad_por_pelicula.id_pelicula
ORDER BY (brechas_edad_por_pelicula.edad_max - brechas_edad_por_pelicula.edad_min) DESC
LIMIT 1;

```

### Consulta 7
**Año de cada premio ganado por la película de mayor duración.** (Si hay empate, elegir solamente uno).


**Ejemplo:**
|Año |
|----|
|2024|
|... |


**Solución:**
Selección de años:
- **Selección de años:** La parte del código SELECT DISTINCT pm.anio selecciona los años de los premios ganados por las películas de mayor duración.
Relación de tablas:
- **Relación de tablas:** Se realiza un JOIN entre la tabla premio_mejor_pelicula (pm) y una subconsulta que encuentra la duración máxima de cada película (max_dur_pelicula). Luego, se relaciona esta subconsulta con la tabla pelicula (p) para encontrar las películas que tienen la duración máxima.
Filtro de resultados:
- **Filtro de resultados:** El JOIN asegura que solo se seleccionen los años de los premios ganados por las películas de mayor duración.
Eliminación de duplicados:
- **Eliminación de duplicados:** Se utiliza DISTINCT para asegurarse de que cada año de premio aparezca solo una vez en los resultados. Esto evita la repetición de años si una película de mayor duración ha ganado premios en múltiples años.


```python
%%sql
SELECT DISTINCT pm.anio
FROM premio_mejor_pelicula pm
JOIN (
    SELECT id_pelicula, MAX(duracion) AS max_duracion
    FROM pelicula
    GROUP BY id_pelicula
) AS max_dur_pelicula ON pm.id_pelicula = max_dur_pelicula.id_pelicula
JOIN pelicula p ON pm.id_pelicula = p.id_pelicula AND p.duracion = max_dur_pelicula.max_duracion;

```

Por supuesto, habrá actores que hayan trabajado en más de una película, con lo que se utiliza DISCTINCT ON para solo obtener una id de película por actor. Por último, a través de la unión con la tabla película y la id de la película, se recupera el nombre de la película y se seleccionan aquellas filas cuyos actores hayan trabajado con más de un director, obteniéndose lo pedido.

```python
%%sql
SELECT tabla2.nombre AS "Nombre del actor", cantidad_director AS "Cantidad de directores", pelicula.nombre AS "Nombre de una película"
FROM
(SELECT DISTINCT ON (actor.id_actor) actor.nombre, cantidad_director, actor_participa_en_pelicula.id_pelicula
FROM(
SELECT actor.id_actor, COUNT(id_director) AS cantidad_director
FROM(
actor_participa_en_pelicula 
LEFT JOIN actor  ON actor.id_actor=actor_participa_en_pelicula.id_actor
LEFT JOIN pelicula ON actor_participa_en_pelicula.id_pelicula=pelicula.id_pelicula)
GROUP BY actor.id_actor
ORDER BY cantidad_director DESC) AS tabla1 LEFT JOIN actor ON tabla1.id_actor=actor.id_actor
LEFT JOIN actor_participa_en_pelicula ON tabla1.id_actor=actor_participa_en_pelicula.id_actor) AS tabla2 LEFT JOIN pelicula
ON tabla2.id_pelicula=pelicula.id_pelicula
WHERE cantidad_director>1
ORDER BY cantidad_director DESC
```

### Consulta 9

**Nombre de todos los actores que han participado sólo en películas dirigidas por el director con id\_director 1.**


**Ejemplo:**
|Nombre      |
|------------|
|Ryan Gosling|
|...         |


**Solución:** 
Filtro por director:
- **Filtro por director:** En la cláusula WHERE, se filtran las películas para seleccionar solo aquellas dirigidas por el director con id_director = 1, lo que se logra con la condición p.id_director = 1.
Relación de tablas:
- **Relación de tablas:** La tabla actor se relaciona con actor_participa_en_pelicula mediante JOIN, utilizando a.id_actor = ap.id_actor. Luego, actor_participa_en_pelicula se relaciona con pelicula también mediante JOIN, utilizando ap.id_pelicula = p.id_pelicula.
Agrupación y conteo:
- **Agrupación y conteo:** Se agrupa el resultado por el nombre del actor (GROUP BY a.nombre) para contar cuántas películas ha dirigido el director con id_director = 1. La cláusula HAVING COUNT(DISTINCT p.id_director) = 1 asegura que solo se seleccionen los actores que han participado en películas dirigidas por un único director (en este caso, el director con id_director = 1).
Selección de columnas:
- **Selección de columnas:** Se selecciona el nombre del actor (a.nombre) para mostrar en los resu-ltados.
sultados.

```python
%%sql
SELECT 
    a.nombre AS nombre_actor
FROM 
    actor a
JOIN 
    actor_participa_en_pelicula ap ON a.id_actor = ap.id_actor
JOIN 
    pelicula p ON ap.id_pelicula = p.id_pelicula
WHERE 
    p.id_director = 1
GROUP BY 
    a.nombre
HAVING 
    COUNT(DISTINCT p.id_director) = 1;

```
